```python
import numpy as np
from matplotlib import pyplot as plt
x = [1,2,3,3,3,3,4,5,5,6,6,6,7,8,9]
plt.hist(x,color='g',bins=4)
plt.show()
```

```python
one =[1,2,3,4,5,6,7,8,9]
two =[1,2,3,4,5,6,3,2,1]
three =[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
one =[1,2,3,4,5,6,7,8,9]
two =[1,2,3,4,5,6,3,2,1]
three =[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]
myexplode = [0.2, 0, 0, 0]
plt.pie(quantity,labels=fruit,autopct ="%0.1f%%",colors=['yellow','grey','blue','green'], explode = myexplode,shadow=True)
plt.legend()
plt.show()
```

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]
plt.pie(quantity,labels=fruit,autopct ="%0.1f%%",colors=['yellow','grey','blue','green'],radius=2)
plt.pie([5],colors=['w'],radius=1)

plt.show()
```
