```python
import numpy as np
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y1=[5,2,3,4,5,16,6,7,9]
y2=[5,2,1,2,3,5,5,8,2]
plt.subplot(1,2,1)
plt.scatter(x,y1,marker="*",color="g",s=100)

plt.subplot(1,2,2)

plt.scatter(x,y2,marker="o",color="r",s=100)

plt.show()
```
