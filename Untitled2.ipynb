```python
import numpy as np
from matplotlib import  pyplot as plt

student = {"Bob":45,"Sam":97,"Matt":23}
names=list(student.keys())
values = list(student.values())
plt.bar(names,values)
plt.title("Marks of student")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
